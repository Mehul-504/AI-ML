```python
df["continent"].unique()
```

```python
total_pop=df["population"].sum()
total_pop
```

```python
overall_life_expectancy = (df["population"]*df["life_expectancy"]).sum()/total_pop
overall_life_expectancy
```

```python
top_10_pop=df.sort_values(by='population',ascending=False).head(10)
top_10_pop
```

```python
df["GDP"] = df["population"]*df["gdp_per_capita"]
df
```

```python
Top_10_GDP=df[df['population'] > 100000000].sort_values(by='gdp_per_capita').head(10)
Top_10_GDP
```

```python
countries_per_cont = df.groupby('continent')['location'].count().reset_index(name='num_of_countries')
countries_per_cont
```

```python
pop_per_continent = df.groupby('continent')['population'].sum().reset_index(name='total_pop')
pop_per_continent
```
