Business insights
- The number of TV shows on Netflix has been increasing at a faster rate than the number of movies in recent years
- The top countries with the most titles on Netflix are the United States, India, and the United Kingdom
- The most common TV rating on Netflix is TV-MA 
- The release of new titles on Netflix has been increasing steadily since 2010, with a sharp increase in 2019 and 2020
- The duration of movies on Netflix varies widely, with some movies being less than an hour long and others over 3 hours long
- The duration of TV shows on Netflix also varies widely, with some shows having only 1 season and others having up to 9 seasons

Recommendations
- Produce more TV shows than movies to cater to the increasing demand for TV shows on Netflix
- Focus on producing content for the top countries with the most titles on Netflix, such as the United States, India, and the United Kingdom
- Continue to produce content with the TV-MA rating, as it is the most common TV rating on Netflix
- Release new titles on Netflix during the months of January, April, and October, as these months have historically had the highest number of new releases
- Consider producing content with a wide range of durations, to cater to different viewer preferences and attention spans
